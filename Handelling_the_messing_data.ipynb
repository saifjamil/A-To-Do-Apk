```python
# Step 4: Handle Missing Values in Numeric Columns Based on Skewness
df = dataset.copy()

print("\n--- Numeric Imputation Based on Skewness ---")
for col in df.select_dtypes(include=['number']):
    skew = df[col].skew()
    method = 'mean' if abs(skew) < 0.3 else 'median'
    value = getattr(df[col], method)()
    df[col].fillna(value, inplace=True)
    print(f"{col}: skew = {skew:.2f} → filled with {method} ({value:.2f})")
```

```python
# Handle non-numeric (categorical) missing values using mode
for col in df.select_dtypes(exclude='number'):
    value = df[col].mode()[0]
    df[col].fillna(value, inplace=True)
    print(f"{col} filled with mode: {value}")
```

```python
missing_data = df.isnull().sum()
print("Missing Data : \n", missing_data)
```
